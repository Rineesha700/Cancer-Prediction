```python
import numpy as np 
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("data.csv")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.drop(['Unnamed: 32', 'id'], axis=1, inplace=True)
```

```python
data.head()
```

```python
# We need to convert target variables into 1s and 0s
```

```python
data.diagnosis =[1 if value == "M" else 0 for value in data.diagnosis]
```

```python
data.head()
```

```python
data['diagnosis'].value_counts().plot(kind='bar', title="Class distributions \n(0: Benign | 1: Malignant)")
```

```python
y = data["diagnosis"] # our target 
X = data.drop(["diagnosis"], axis=1) # our features
```

```python
y
```

```python
X
```

```python
# Normalize/scale the data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
# Now we are gonna go train-test-split
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.30, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
# Evaluate the model
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```
